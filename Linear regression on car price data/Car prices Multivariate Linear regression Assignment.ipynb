```python
#We can divide the companies into buckets of low, med and high mased on the median price of that company. 
mean_dict = prices.groupby(['company'])[['price']].median().to_dict()
mean_dict = mean_dict['price']
mean_dict
```

```python
dict_keys = list(mean_dict.keys())

# Median price of category below 10000 is low, between 10000 and 20000 is med and above 20000 is high
for i in dict_keys:
    if mean_dict[i] < 10000:
        mean_dict[i] = 'low'
    elif mean_dict[i] >= 10000 and mean_dict[i] <= 20000:
        mean_dict[i] = 'med'
    else:
        mean_dict[i] = 'high'

prices.company = prices.company.map(mean_dict)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = list(df_train.columns)
df_train[cols] = scaler.fit_transform(df_train[cols])
```

```python
y_train = df_train.pop('price')
x_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
x_train = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train).fit()
```

```python
fig , ax = plt.subplots(1,3, figsize = (15,5))
plt.suptitle('Model analysis Testing')
ax[0].set_title('Error Distribution')
p1 = sns.histplot(residuals_test, kde = True, bins = 50, ax = ax[0])
ax[1].set_title('Actual vs Predictions')
p2 = sns.scatterplot(y_test,y_predictions_test, ax = ax[1])
ax[2].set_title('Residual Distribution around mean')
p3 = sns.scatterplot(y_test, residuals_test, ax = ax[2])
mean = sns.lineplot(y_test, [residuals_test.mean()]*len(residuals_test), ax = ax[2])
plt.tight_layout()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
rmse = sqrt(mean_squared_error(y_test, y_predictions_test))
print('Model RMSE:',rmse)

r2=r2_score(y_test, y_predictions_test)
print('Model r2_score:',r2)
```

```python
from statsmodels.stats.outliers_influence  import  variance_inflation_factor 
def get_VIF(X_train):
 # A dataframe that will contain the names of all the feature variables and their respective VIFs 
    vif = pd.DataFrame() 
    vif['Features'] = X_train.columns 
    vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])] 
    #vif['VIF'] = (vif['VIF'], 2) 
    vif = vif.sort_values(by = "VIF", ascending = False) 
    return (vif) 
```
